```python
def differentiate_variables(data):
    continuous = []
    discrete_vars = []

    for column in data.columns:
        unique_values = data[column].nunique()
        if unique_values > 10:  # Adjust the threshold as needed
            continuous.append(column)
        else:
            discrete_vars.append(column)

    continuous_df = data[continuous]
    discrete_df = data[discrete_vars]

    return continuous_df, discrete_df
```

```python
differentiate_variables(data)
```
